```python
print("You have entered the Trifecta.")
input("Please proceed forward: ")

inventory = []

def ask_to_continue():
    ans = input("Continue? (yes/no): ").lower()
    if ans == "yes":
        return True
    elif ans == "no":
        print("Ok. Goodbye.")
        exit()
    else:
        print("Invalid input. Please enter 'yes' or 'no'.")
        return ask_to_continue()

def firstroom():
    OPTIONS = ["Enter hall", "Leave", "Take candles", "inventory"]
    while True:
        print("\nThe First room consists of a vast dark dining room.\nA long table is lit with candles as you approach.")
        print(OPTIONS)
        ans = input("What would you like to do? ").lower()

        if ans == "enter hall":
            secondroom()
        elif ans == "leave":
            print("Sorry, but you cannot leave the trifecta this way.")
            if ask_to_continue():
                continue
                firstroom()
        elif ans == "take candles":
          if "candles" in inventory:
            print("You already have the candles.")
            if ask_to_continue():
              continue
          else:
            inventory.append("candles")
            print("You have taken the candles.")
            if ask_to_continue():
              continue
        elif ans == "inventory":
            view_inventory()
        else:
            print("Invalid option. Please try again.")

        if not ask_to_continue():
            break

def secondroom():
    OPTIONS = ["Enter third room", "inventory", "go back"]
    while True:
        print("\nThe second room is a dark hallway with paintings that seem... off.")
        print(OPTIONS)
        ans = input("What would you like to do? ").lower()

        if ans == "enter third room":
            thirdroom()
        elif ans == "inventory":
            view_inventory()
        elif ans == "go back":
            print("Would you like to go back to the first room?")
            if ask_to_continue():
                firstroom()
        elif ans == "paintings":
            print("The paintings seem burnable.")
            burn_ans = input("Burn it? (yes/no): ").lower()
            if burn_ans == "yes" and "candles" in inventory:
                print("Secret Found!\nYou have burned the paintings and found a key!")
                inventory.append("key")
                if ask_to_continue():
                    continue
            elif burn_ans == "yes" and "candles" not in inventory:
                print("You cannot burn the paintings without candles.")
            else:
                print("You cannot burn the paintings without candles.")
                if ask_to_continue():
                  continue
        else:
            print("Invalid option. Please try again.")

        if not ask_to_continue():
            break

def thirdroom():
    OPTIONS = ["Exit", "inventory", "Steal jewels"]
    while True:
        print("\nThe third room is a treasurous area with jewels and a chest.\nThere is an exit door in front of you.")
        print(OPTIONS)
        ans = input("What would you like to do? ").lower()

        if ans == "exit":
            print("You cannot escape the trifecta in this way, you will be brought back to the beginning.")
            if ask_to_continue():
                continue
            firstroom()
        elif ans == "inventory":
            view_inventory()
        elif ans == "steal jewels":
            print("You greedy mole, you cannot steal the jewels.\nYou will be killed for your mistake.\nYou died.")
            inventory.clear()
            if ask_to_continue():
              firstroom()
        elif ans == "chest":
            unlock_ans = input("Secret Found!\nThe chest is locked. Type 'unlock' to unlock it: ").lower()
            if unlock_ans == "unlock" and "key" in inventory:
                drink_ans = input("You have unlocked the chest and found a potion! Drink it? (yes/no): ").lower()
                if drink_ans == "yes":
                    print("You drift into a peaceful sleep, and wake up with your cat licking your face. It was all a dr3am.....3333.33.33.333333\nCongratulations, you have escaped the trifecta.")
                    break
                else:
                    print("You left the potion.")
                    if ask_to_continue():
                        continue
            elif unlock_ans == "unlock" and "key" not in inventory:
                print("You cannot unlock the chest without the key.")
            else:
                print("What?")
        else:
            print("Invalid option. Please try again.")

        if not ask_to_continue():
            break

def view_inventory():
    if len(inventory) == 0:
        print("Nothing in inventory.")
    else:
        print("Inventory:", ", ".join(inventory))

firstroom()
```
